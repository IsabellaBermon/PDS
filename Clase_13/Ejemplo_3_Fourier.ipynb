# Ejemplo 3. Transformada de Fourier.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fourier
```

### Transformada de Fourier de una se√±al cuadrada

```python
x = np.concatenate([[0 for i in range(200)], [1 for i in range(7)], [0 for i in range(200)]])
N = x.size
n = np.arange(-N//2+1, N//2+1)
```

```python
plt.figure()
plt.title('Se√±al $x(n)$', fontsize=18)
plt.stem(n, x)
plt.xlabel('n')
plt.xlim([-7, 7])
plt.grid()
plt.show()
```

```python
X = fourier.fft(x)
```

```python
X_Mag = abs(X)
X_ang = np.angle(X)
```

```python
plt.figure()
plt.title('Espectro $X(\omega)$', fontsize=18)
plt.plot(np.linspace(-np.pi, np.pi, len(X_Mag)), np.fft.fftshift(X_Mag))
plt.xlabel('$\omega$')
plt.grid()
plt.show()
```
