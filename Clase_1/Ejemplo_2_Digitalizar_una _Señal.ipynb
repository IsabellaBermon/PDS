# Ejemplo 2. Digitalización de señales analógicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grafica(signal):
    plt.figure(figsize=(20,5))
    plt.plot(t, signal, 'b', label = 'Señal analógica')
    plt.stem(t, signal, linefmt='r', label='Digital signal')
    plt.xlabel('Tiempo')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()
    plt.show()
```

## Señal sinusoidal

```python
# Definimos el período de la señal
T = 2*np.pi
# Definimos la frecuencia de muestreo de la señal
fs = 40
# Definimos el tiempo de muestreo y el número de muestras
t = np.linspace(0, T, fs) # Vector de tiempo
signal = np.sin(t)        # Señal analogica
# Grafica de la señal analógica y digital
grafica(signal)
```

## Señal rectangular

```python
# Definimos la frecuencia de la señal
f = 1 # Hz 
# Definimos el período de la señal
T = 1/f
# Definimos el tiempo de muestreo y el número de muestras
t = np.linspace(0, 2*T, 100) # el número de muestras es mucho mayor a su frecuencia para mejor representación
# Definimos la señal rectangular
rect = np.zeros_like(t)
rect[t<=T/2] = 1
rect[t>T/2] = 0
# Grafica de la señal analógica y digital
grafica(rect)
```

## Señal hiperbólica

```python
# Definimos el vector de tiempo
t = np.linspace(-5, 5, 200)
# Definimos la señal hiperbólica
y = np.tanh(t)
# Graficamos la señal analógica y digital
grafica(y)
```
