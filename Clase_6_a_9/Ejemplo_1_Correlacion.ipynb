# Ejemplo 1. Correlaci칩n.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 4, 2, 5, 1])
y = np.array([4, 2, 5])
Rxy  = np.correlate(x, y, mode='full')
```

```python
# Grafica de la correlaci칩n
plt.figure()
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplo 1.2 Correlaci칩n de se침al polinomial.

```python
coef = [1, 2, -3, 2] 
x = np.linspace(0, 1, 10)
y = np.polyval(coef, x)

# Grafica de la secuencia
plt.figure(figsize=(10,4))
plt.stem(x, y)
plt.title('Funci칩n polinomial')
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

Rxy  = np.correlate(x, y, mode='full')

# Grafica de la correlaci칩n
plt.figure(figsize=(10,4))
plt.stem(Rxy)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
