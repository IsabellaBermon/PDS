# Ejemplo 1. Diseño de Filtros IRR.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
from numpy import pi
```

![Imagen](Img/Ha.png)

Se puede dar una función en particular, en caso contrario se parte de la función de transferencia del filtro analógico.

Se quiere diseñar un filtro digital con una frecuencia de corte de 0.125 Hz, o sea, 0.25 rad/s.

$\Omega_{c}=\frac{2tan(\frac{0.25\pi}{2})}{T_S}=\frac{0.828}{T_S}$

$H_{a}(s)=\frac{1}{1+\frac{T_{S}S}{0.282}}$

![Imagen](Img/bilineal.png)

Haciendo los calculos necesarios se llega a que el filtro digital es de la siguiente forma:

$H(Z)=\frac{0.292+0.292Z^{-1}}{1-0.4159Z^{-1}}$

```python
b = [0.292, 0.292]
a = [1, -0.159]
w, h = sp.freqz(b, a)
```

```python
fig = plt.figure()
plt.title('Digital filter frequency response')
ax1 = fig.add_subplot(111)
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
plt.plot(w, angles, 'g')
plt.ylabel('Angle (radians)', color='g')
plt.grid()
plt.axis('tight')
plt.show()
```

En conclusión, para diseñar un filtro digital a partir de un filtro anlógico es necesario hacer el proceso analítico para llegar a una expresión para el filtro digital que permita diferenciar correctamente el numerador y denominador termino a termino, una vez se tiene esto se aplica la funcion sp.freqz() de la forma en que se muestra, siendo 'b' el numerador y 'a' el denominador.
