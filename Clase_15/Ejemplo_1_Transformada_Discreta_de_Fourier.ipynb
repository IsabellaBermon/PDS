# Ejemplo 1. Transformada Discreta de Fourier.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
import scipy.fftpack as fourier
```

### $x(n)=0.8^n u(n)$

```python
N = 50
n = np.arange(N)
x_n = 0.8**n
```

```python
X = fourier.fft(x_n)
```

```python
X_Mag = abs(X)
X_Mag = X_Mag/max(X_Mag) # Normalizaci√≥n
X_ang = np.angle(X)
```

```python
plt.subplot(211)
plt.stem(n, x_n, label='x(n)')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()

plt.subplot(212)
plt.stem(np.linspace(-np.pi, np.pi, len(X_Mag)), X_Mag, label='X(w)')
plt.xlabel('w')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```
