# Ejemplo 2. Convoluci칩n de Se침ales.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

$x(n)=[\overrightarrow{4}, 3, 2]$

$h(n)=[10, \overrightarrow{20}, 10]$

```python
# Sabiendo que h(n) tiene el cero en el 20 y x(n) lo tiene en el primer elemento:
# N1=0 y N2=-1. M1=2, M2=1. N1+N2 <= n <= M1+M2
# El rango debe ser desde [-1,3]
n = np.arange(-1,4)
h = np.array([10, 20, 10])  # h(n)
x = np.array([4, 3, 2])     # x(n)
y = np.convolve(h, x)       # Convoluci칩n de las se침ales y(n)
```

```python
# Grafica de la convoluci칩n
plt.figure()
plt.stem(n, y)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

Otro ejemplo de convoluci칩n:

$j(n)=[\overrightarrow{-2}, 0, 1, -1, 3]$

$k(n)=[\overrightarrow{1}, 2, 0, -1]$

```python
# En este caso los ceros de ambos arreglos est치n en el primer elemento
# entonces se sabe que N1=N2=0 y M1=4, M2=3
# El rango es entonces: 0 <= n <= 7
n2 =np.arange(0, 8)
j = np.array([-2, 0, 1, -1, 3])
k = np.array([1, 2, 0, -1])
z = np.convolve(j, k)
```

```python
# Grafica de la convoluci칩n
plt.figure()
plt.stem(n2, z)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
