# Ejemplo 1. Linealidad e Invarianza Temporal

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def desplazar_nper(secuencia, desplazamiento):
    # Toma el desplazamiento, y, mientras el contador sea menor a ese contadorvan a haber ceros,
    # luego de eso, se usa el mismo desplazamiento para agarrar los valores que se omitieron por la
    # sentencia IF.
    secuencia_despl = []
    for i in range(len(secuencia)):
        if (i < desplazamiento):
            secuencia_despl.append(0)
        else:
            secuencia_despl.append(secuencia[i-desplazamiento])
    return secuencia_despl
```

```python
t = np.arange(0, 10)
x = lambda k: np.exp(-k) # Prototipo de la función a la que se le aplican las transformaciones
x1 = 2*x(t)
x2 = desplazar_nper(3*x(t), 1)
x3 = desplazar_nper(x(t), 2)

y = x1+x2+x3
```

```python
# Grafica de la señal de salida y
plt.figure()
plt.stem(t, y)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.ylim([0, 4])
plt.grid()
plt.show()
```

# Ejemplo 1.2 Linealidad e Invarianza Temporal

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define system
def system(x, k):
    return k * x

# Create input signals
n = np.arange(-20, 20)

x1 = np.sin(2 * np.pi * n / 10)
x2 = np.cos(2 * np.pi * n / 10)
x3 = x1 + x2

# Apply system to input signals
k = 2
y1 = system(x1, k)
y2 = system(x2, k)
y3 = system(x3, k)

plt.figure(figsize=(14, 4))
plt.stem(n, x1, label='x1',linefmt='r:')
plt.stem(n, x2, label='x2')
plt.title('Señales de entreda')
plt.legend()
plt.show()

plt.figure(figsize=(14, 4))
plt.stem(n, y1, label='y1',linefmt='r:')
plt.stem(n, y2, label='y2')
plt.title('Señales de salida')
plt.legend()
plt.show()

plt.figure(figsize=(14, 4))
plt.stem(n, y3, label='y3')
plt.title('Linealidad')
plt.legend()
plt.show()

x4 = np.sin(2 * np.pi * (n - 5) / 10)
y4 = system(x4, k)
plt.figure(figsize=(14, 4))
plt.stem(n, x4, label='x4')
plt.title('Invarianza')
plt.legend(loc='best')
plt.show()
```
