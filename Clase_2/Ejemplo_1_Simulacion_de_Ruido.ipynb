# Ejemplo 1. Como simular ruido blanco a una señal analógica.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grafica(signal):
    plt.figure(figsize=(20,5))
    plt.plot(t, signal, 'b', label = 'Señal analógica')
    plt.stem(t, signal, linefmt='r', label='Digital signal')
    plt.xlabel('Tiempo')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 20)     # Vector de tiempo
signal = np.sin(t)                  # Señal
noise = np.random.randn(20)         # ruido blanco
signal_with_noise = signal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal son ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

# Ejemplo 2. Como simular ruido Gaussiano a una señal analógica.

```python
# Definimos el período de la señal
T = 2*np.pi
# Definimos la frecuencia de muestreo de la señal
fs = 40
# Definimos el tiempo de muestreo y el número de muestras
t = np.linspace(0, T, fs) # Vector de tiempo
signal = np.sin(t)        # Señal analogica
# Grafica de la señal original
grafica(signal)
# Se añade el ruido
media = 0
desviacion = 0.5
noise_signal=np.random.normal(media,desviacion,len(signal))
signal_with_noise = signal + noise_signal  # Se añade el ruido a la señal original
# Grafica de la señal de ruido Gaussiano con media 0 y desviación estándar 0.5 
grafica(signal_with_noise)
```

# Ejemplo 3. Como simular ruido de cuantización a una señal analógica.

```python
grafica(signal)
# Simular ruido de cuantización
bits = 4  # Número de bits de cuantización
levels = 2**bits  # Número de niveles de cuantización
delta = (np.max(signal) - np.min(signal)) / levels  # Tamaño del intervalo de cuantización
quantized_signal = np.round(signal / delta) * delta
grafica(quantized_signal)
```
