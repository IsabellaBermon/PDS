# Ejemplo 1. Filtros Digitales.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp
```

Como ejemplo se va a diseñar un filtro analogico pasa bajos. Posteriormente se va a explicar a detalle como se diseñan filtros de forma analítica. Vamos a definir las tres partes fundamentales a considerar a la hora de diseñar un filtro ANCHO DE BANDA DE TRANSICIÓN, RIZADO y ATENUACIÓN.

```python
b = sp.firwin(80, 0.5, window=('kaiser', 8))
w, h = sp.freqz(b, 1, whole=True)
plt.figure()
plt.title('Digital filter frequency response')
plt.plot(w-np.pi, np.abs(np.fft.fftshift(h)), 'b')
plt.xlim([0, np.pi])
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
plt.show()
```

El anterior es un filtro FIR, para ver como varia el ANCHO DE BANDA DE TRANSICIÓN del filtro lo que se hace es variar el orden del filtro. Si se disminuye el orden del filtro, aumenta el ancho de banda de transición.

```python
b = sp.firwin(20, 0.5, window=('kaiser', 8)) # Aqui esta el cambio, antes era sp.firwin(80, 0.5...
w, h = sp.freqz(b, 1, whole=True)
plt.figure()
plt.title('Digital filter frequency response')
plt.plot(w-np.pi, np.abs(np.fft.fftshift(h)), 'b')
plt.xlim([0, np.pi])
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
plt.show()
```

Se evidencia como la caida es menos pronunciada, esto indica que el ancho de banda de transición aumentó mucho con respecto al anterior.

Ahora se va a mostrar como varía el rizado del filtro, esto se hace con el tipo de enventanado, mas adelante se verá como escoger la ventana adecuada para obtener un rizado determinado.

```python
b = sp.firwin(80, 0.5, window=('kaiser', 1)) # Aqui esta el cambio, antes era window=('kaiser', 8)
w, h = sp.freqz(b, 1, whole=True)
plt.figure()
plt.title('Digital filter frequency response')
plt.plot(w-np.pi, np.abs(np.fft.fftshift(h)), 'b')
plt.xlim([0, np.pi])
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
plt.show()
```

Se evidencia que aparecio rizado al modificar la ventana existente.

Por último, pero no menos importante, se va a mostrar como sacar la grafica de la atenuación del filtro.

```python
b = sp.firwin(80, 0.5, window=('kaiser', 8))
w, h = sp.freqz(b)
plt.figure()
plt.title('Digital filter frequency response')
plt.plot(w, 20 * np.log10(abs(h)), 'b') # Aqui está el cambio
plt.ylabel('Amplitude [dB]', color='b')
plt.xlabel('Frequency [rad/sample]')
plt.show()
```
