# Ejemplo2. Generar señales útiles como el delta, escalón, y rampa.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

Es muy importante tener bien definido el vector de muestras, este se genera en base N que es el numero de muestras, se recomienda siempre un N impar.

```python
N = 9
n = np.arange(-N//2+1, N//2+1)
```

#### 1. Función delta

```python
delta = np.zeros(N) # Vector de ceros
delta[N//2] = 1     # Se pone un uno en la mitad

# Grafica de la secuencia
plt.figure()
plt.title('Función delta')
plt.stem(n, delta)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 2. Función escalón unitario

```python
amplitud_escalon = 1    # Amplitud
desplazamiento = 1      # Desplazamiento del escalón
u = np.heaviside(n-desplazamiento, amplitud_escalon)    # Función escalón

# Otra forma de hacerlo:
# u = np.piecewise(n, n>=1, [1, 0])

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función escalón unitario')
plt.stem(n, u)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 3. Función rampa

```python
rampa = []              # Vector vacío para rellenar
desplazamiento2 = 0     # Desplazamiento de la rampa
cont = 1                # Contador para generar la rampa
for i in range(N):
    if (i >= (N//2+1)-(-desplazamiento2)):
        rampa.append(cont)
        cont += 1
    else:
        rampa.append(0)

# Grafica de la secuencia
plt.figure()
plt.figure()
plt.title('Función rampa')
plt.stem(n, rampa)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 4. Función polinomial ( Dos formas para definir el polinomio )

```python
# Definir los coeficientes del polinomio (de mayor a menor grado)
coef = [1, 2, -3, 2] # ^3 ^2 ^1 ^0
# Generar 10 muestras espaciadas entre 0 y 1
x = np.linspace(0, 1, 10)
# Evaluar el polinomio en los valores generados
y = np.polyval(coef, x)

# Grafica de la secuencia
plt.stem(x, y)
plt.title('Función polinomial')
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

```python
a = 8 # Coeficiente cuadrático
b = 2 # Coeficiente lineal
c = 15 # Término independiente
n = np.arange(N)
polinomio = a*(n**2) + b*n + c

# Grafica de la secuencia
plt.figure()
plt.title('Función polinómica')
plt.stem(n, polinomio)
plt.xlabel('Número de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

#### 5. Función exponencial

```python
amplitud_exp = 1 # Amplitud
tau = 5 # Constante de tiempo
desplazamiento = 2 # Desplazamiento de la exponencial

n = np.arange(20) # vector de muestras
# Evaluación de la función exponencial en cada muestra
exp = amplitud_exp * np.exp(-(n-desplazamiento)/tau)

# Graficación de la secuencia
plt.figure()
plt.title('Función exponencial')
plt.stem(n, exp)
plt.xlabel('Número de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
