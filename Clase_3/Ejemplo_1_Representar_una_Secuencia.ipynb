# Ejemplo 1. Representar una secuencia.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_n = [1, 4, 3, 0, 4, 2, 3, 4, 0, 1, 4] # Secuencia
n = np.arange(len(x_n))                 # Vector de muestras
```

```python
# Grafica de la secuencia
plt.figure()
plt.stem(n, x_n)
plt.xlabel('Numero de muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

# Ejemplo 2. Representar una secuencia aleatoria, invertirla y desplazarla

```python
import numpy as np 
import matplotlib.pyplot as plt

# Generar una secuencia aleatoria de 10 muestras
seq_aleatoria = np.random.rand(10)

# Graficar la secuencia aleatoria
plt.stem(seq_aleatoria)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Secuencia aleatoria de 10 muestras')
plt.grid()
plt.show()
```

```python
# Invertir la secuencia
seq_aleatoria_invertida = np.flip(seq_aleatoria)

# Grafica de la secuencia
plt.stem(seq_aleatoria_invertida)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Secuencia aleatoria de 10 muestras invertida en el eje x')
plt.grid()
plt.show()
```

```python
# Rotar la secuencia en dos posiciones
n = 1
seq_aleatoria_rotada = np.roll(seq_aleatoria_invertida, n)

# Graficar la secuencia rotada
plt.stem(seq_aleatoria_rotada)
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Secuencia aleatoria de 10 muestras rotada en x-n')
plt.grid()
plt.show()
```
