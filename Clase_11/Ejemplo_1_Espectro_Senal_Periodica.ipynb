# Ejemplo 1. Cálculo del Espectro de una Eeñal Periódica

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 15
n = np.arange(N)
x = np.cos(2*np.pi*n/3) + np.sin(2*np.pi*n/5)

plt.figure()
plt.stem(n, x)
plt.ylabel('Amplitud')
plt.xlabel('Numero de muestras')
plt.grid()
plt.show()
```

```python
def fourier_coefs(x_n, Ncoef):
    # x n: senal definida en un periodo
    # Ncoef: numero de componentes a hallar
    N = x_n.size
    cn = np.empty(Ncoef, 'complex_')
    n = np.arange(N)
    for k in np.arange(Ncoef//2+1):
        cn[k] = (1/N) * np.sum(x_n*np.exp(-1j*2*np.pi*k*n/N)) # Calcular la parte positiva de los coef.
        cn[-k] = (1/N) * np.sum(x_n*np.exp(-1j*2*np.pi*-k*n/N)) # Calcular la parte negativa de los coef.
    return cn
```

```python
Ck = fourier_coefs(x, N)

for i, ck in enumerate(Ck):
    print(i, ':', ck)
```

```python
espectro = np.abs(Ck)**2
k = np.arange(-N//2+1, N//2+1)

plt.figure()
plt.stem(k, espectro[k])
plt.ylabel('$|C_k|^2$')
plt.xlabel('k')
plt.show()
```
